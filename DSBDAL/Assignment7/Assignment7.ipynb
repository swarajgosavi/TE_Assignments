## 7.3 Representation of Document

### 7.3.1 Constructing Corpus

```python
corpus = [ text, text2 ]
corpus
```

```python
# create a word set for the corpus
# Similar to word_tokens
words_set = set()

for doc in  corpus:
    words = doc.split()
    words_set = words_set.union(set(words))
    
print('Number of words in the corpus:',len(words_set))
print('The words in the corpus: \n', words_set)
```

### 7.3.2 Trem Frequency
#### Computing Term Frequency

**Term Frequency:**
TF of a term or word is the number of times the term appears in a document compared to the total number of words in the document.

\begin{equation}
TF = \frac { number\ of\ times\ the\ trem\ appers\ in\ a\ document } { total\ number\ of\ words\ in\ the\ document }
\end{equation}

```python
n_docs = len(corpus)         # Number of documents in the corpus
n_words_set = len(words_set) # Number of unique words in the corpus

df_tf = pd.DataFrame(np.zeros((n_docs, n_words_set)), columns=list(words_set))

# Compute Term Frequency (TF)
for i in range(n_docs):
    words = corpus[i].split() # Words in the document
    for w in words:
        df_tf[w][i] = df_tf[w][i] + ( 1 / len(words))

df_tf.head()
```

```python
def termfreq(corpus):
    dic = {}

    for doc in corpus:
        for word in doc.split():
            if word in dic:
                dic[word] += 1
            else:
                dic[word] = 1

    for word, freq in dic.items():
        print(word, freq)
        dic[word] = freq / len(doc)

    print("Document size in no of words: ", len(doc))

    return dic

termfreq(corpus)
```

### 7.3.3 Inverse Document Frequency
#### Computing Inverse Document Frequency

**Inverse Document Frequency:** IDF of a term reflects the proportion of documents in the corpus that contain the term. Words unique to a small percentage of documents (e.g., technical jargon terms) receive higher importance values than words common across all documents (e.g., a, the, and).

\begin{equation}
IDF = log( \frac { number\ of\ documents\ in\ the\ corpus } { number\ of\ documents\ in\ the\ corpus\ contains\ the\ term } )
\end{equation}

**Note** that there are some different approaches to calculating the IDF score. The base 10 logarithm is often used in the calculation. However, some libraries use a natural logarithm. In addition, one can be added to the denominator as follows in order to avoid division by zero.

\begin{equation}
IDF = log( \frac { number\ of\ documents\ in\ the\ corpus } { { number\ of\ documents\ in\ the\ corpus\ contains\ the\ term } + 1 } )
\end{equation}

```python
print("IDF of: ")

idf = {}

for w in words_set:
    k = 0 # nnumber of documnets in the corpus that contains this word

    for i in range(n_docs):
        if w in corpus[i].split():
            k += 1

    idf[w] = np.log10(n_docs / k)

    print(f'{w:>15}: {idf[w]:>10}')
```

#### Putting it Together: Computing TF-IDF

The TF-IDF of a term is calculated by multiplying TF and IDF scores.


$$
TF-IDF = TF * IDF
$$

```python
df_tf_idf = df_tf.copy()

for w in words_set:
    for i in range(n_docs):
        df_tf_idf[w][i] = df_tf[w][i] * idf[w]
        
df_tf_idf
```
