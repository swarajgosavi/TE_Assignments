```python
%matplotlib inline

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

#### Import the Data
Let's extract the our Titanic data from the .csv file, create a pandas DataFrame and look at the available indicators:

- **Survived:** Outcome of survival (0 = No; 1 = Yes)
- **Pclass:** Socio-economic class (1 = Upper class; 2 = Middle class; 3 = Lower class)
- **Sex:** Sex of the passenger
- **Age:** Age of the passenger (Some entries contain NaN)
- **SibSp:** Number of siblings and spouses of the passenger aboard
- **Parch:** Number of parents and children of the passenger aboard
- **Fare:** Fare paid by the passenger
- **deck:** Deck number of the passenger (Some entries contain NaN)
- **Embarked:** Port of embarkation of the passenger (C = Cherbourg; Q = Queenstown; S = Southampton)

```python
#importing dataset
data = sns.load_dataset("titanic")
```

### Line Plot

Observation: Females survived the most, than males. Also Childerns and elderly person survived in camparison to age group of 20 to 40 years.

```python
plt.figure(figsize=(8,8))
sns.lineplot(x="age", y="survived", hue="sex", style="sex", data=data)
plt.show()
```

### Scatter Plot

Observation: Most of the passenger has fare below 100 while some outliers with higher fare as 500

```python
plt.figure(figsize=(8,8))
sns.scatterplot(x="age", y="fare", hue="sex", data=data)
plt.show()
```

### Box Plot

Observation: The distribution of the age of passengers who survived appears to have a smaller median than that of the passengers who didn't survived, indicating younger passengers survived most

```python
plt.figure(figsize=(8,8))
sns.boxplot(x="survived", y="age", data=data)
plt.show()
```

### Point Plot

Observation: Females of 1st and 2nd class has survived in camparision to males og all 3 class

```python
plt.subplots(figsize=(8, 8))
sns.pointplot(x="sex", y="survived", hue="pclass", data=data)
plt.show()
```

### Count plot

Observation: Pclass = 3 (third class) passengers survived the least and Pclass = 1 (first class) passengers survived the most

```python
sns.countplot(x="pclass", hue="survived", data=data);
plt.show()
```

### Violin plot

```python
sns.violinplot(x="survived", y="age", hue='sex', data=data)
plt.show()
```

### Swarm plot

```python
plt.figure(figsize=(8,8))
sns.swarmplot(x="sex", y="age", data=data)
plt.show()
```

### Bar plot

```python
plt.figure(figsize=(8,8))
sns.barplot(x="sex", y="survived", hue="pclass", data=data)
plt.show()
```

### KDE Plot

```python
sns.kdeplot(data['age'])
plt.show()
```

```python
sns.histplot(x="fare", data=data, bins=100)
plt.show()
```

```python
sns.histplot(x="fare", data=data, bins=10)
plt.show()
```

Observation: As first class passengers survived the most, passengers with higher fare survived, assuming the fact that first-class had a higher fare than other classes
